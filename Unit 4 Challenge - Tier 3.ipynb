**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:

```python
#Solving the brief without using the function
import locale

# Set the locale to the UK for formatting as English pounds
locale.setlocale(locale.LC_ALL, 'en_GB.UTF-8')

# Filter the dataframe for the years 2018 and 1998
filtered_df = london_boroughs_df[london_boroughs_df['Year'].isin([2018, 1998])]

# Calculate the average price for each borough for the years 2018 and 1998
avg_price_df = filtered_df.groupby(['Borough', 'Year'])['Avg_Price'].mean().reset_index()

# Pivot the dataframe to have the years as columns
pivot_df = avg_price_df.pivot(index='Borough', columns='Year', values='Avg_Price').reset_index()

# Rename the columns for clarity
pivot_df.columns = ['Borough', 'Avg_Price_1998', 'Avg_Price_2018']

#Add price ratio column to the df and convert to a percentage
pivot_df['Price_Ratio'] = pivot_df['Avg_Price_1998'] / pivot_df['Avg_Price_2018']
pivot_df['Price_Ratio'] = pivot_df['Price_Ratio'] * 100

# Format the prices to two decimal places
pivot_df['Avg_Price_1998'] = pivot_df['Avg_Price_1998'].apply(lambda x: locale.currency(x, grouping=True))
pivot_df['Avg_Price_2018'] = pivot_df['Avg_Price_2018'].apply(lambda x: locale.currency(x, grouping=True))

# Sort the dataframe by Price_Ratio in descending order
pivot_df = pivot_df.sort_values('Price_Ratio', ascending=False)

pivot_df
```

```python
#Attempting the function

def create_price_ratio(london_boroughs_df, borough_filter):
    # Creates a new dataframe that contains only the data for the specified borough
    borough_data2 = london_boroughs_df[london_boroughs_df['Borough'] == borough_filter]
    
    # Get the average price for the years 1998 and 2018
    price_1998 = borough_data2.loc[borough_data2['Year'] == 1998, 'Avg_Price'].mean()
    price_2018 = borough_data2.loc[borough_data2['Year'] == 2018, 'Avg_Price'].mean()

    # Calculate the price ratio
    ratio = price_1998 / price_2018

    return ratio    

create_price_ratio(london_boroughs_df, 'Harrow')
```

```python
#Loop to iterate through all boroughs and creates a dictionary called price_ratio 
unique_boroughs = london_boroughs_df['Borough'].unique()

price_ratios = {}

for borough in unique_boroughs:
    ratio = create_price_ratio(london_boroughs_df, borough)
    price_ratios[borough] = ratio
```

```python
#Convert the price_ratio dictionary to a data frame
df_price_ratios = pd.DataFrame.from_dict(price_ratios, orient='index', columns=['Price_Ratio'])

#Sort the data frame to have the highest price_ratios on top
df_price_ratios_sorted = df_price_ratios.sort_values(by='Price_Ratio', ascending=False)

df_price_ratios_sorted['Price_Ratio'] = df_price_ratios_sorted['Price_Ratio'] * 100

df_price_ratios_sorted
```
